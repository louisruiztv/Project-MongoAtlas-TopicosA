# Update Documents and Collections

```python
from connect import db

modificar = db.libros

res = modificar.update_one(
    {"Titulo": "La ciudad verde"},   # filtro
    {"$inc": {"Stock": 10}}          # operador $inc con valor numérico
)

print("Documentos modificados:", res.modified_count)

for doc in modificar.find({"Titulo": "La ciudad verde"}):
    print(doc)
```

```python
for doc in modificar.find({"Titulo": "La ciudad verde"}):
    print(doc['Titulo'], "ahora hay ", doc['Stock'])
```

```python
# Actuallizar la disponiblidad a true de todos usando update_many
from connect import db
modificar = db.libros
res = modificar.update_many(
    {},                              # filtro (todos los documentos)
    {"$set": {"prestado": True}}   # operador $set con valor booleano
)
print("Documentos modificados:", res.modified_count)
```

```python
#BUlK WRITE OPERATIONS
from pymongo import UpdateOne
from connect import db
operations = [
    UpdateOne(
        {"Titulo": "La ciudad verde"},     # filtro
        {"$set": {"Stock": 6}}           # actualización
    ),
    UpdateOne(
        {"Titulo": "La ciudad verde"},
        {"$inc": {"Stock": 2}}
    )
]

# Ejecutar todas las operaciones en bloque
result = modificar.bulk_write(operations)

print("Documentos modificados:", result.modified_count)
```
