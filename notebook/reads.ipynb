# Realizando consultas

```python
from connect import db
from bson.objectid import ObjectId # para mostrar en forma correcta los IDs

resultados = db.libros

# Buscar todos los documentos
for doc in resultados.find():
    print(doc)
```

```python
# Buscar por autor
for doc in resultados.find({"autor": "Gabriel García Márquez"}):
    print(doc['titulo'], "-", doc['autor'])
```

```python
# Consulta muesta libros mayore tal anio 1993
for doc in resultados.find({"anio": {"$gt": 1960}}):
    print(doc['titulo'], "-", doc['anio'])
```

```python
# Consulta muestra libros entre 1990 y 2000
for doc in resultados.find({"anio": {"$gte": 1980, "$lte": 2000}}):
    print(doc['titulo'], "-", doc['anio'])
```

```python
#Listar los libro de ciencia ficcion
for doc in resultados.find({"generos": "ciencia ficción"}):
    print(doc['titulo'], "-", doc['generos'])
```

```python
# Libros con stock mayora 2
for doc in resultados.find({"Stock": {"$gt": 2}}):
    print(doc['Titulo'], doc['Anio'], "Stock:", doc['Stock'])
```
